```python
import warnings
warnings.filterwarnings("ignore")

import os
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"  # disable GPU usage

import numpy as np
import pandas as pd
# import tensorflow as tf
# from scipy.optimize import curve_fit
from matplotlib import pyplot as plt
from tensorflow import keras

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

from sklearn.externals import joblib
```

```python
data_path = "../experimental_data/data_median.csv"
model_path = "network_median_model.h5"
scaler_path = "median_scaler_temp.save"
figs_path = "figs_median/median"
```

```python
data = pd.read_csv(data_path)
print(data.shape)
data.head()
```

```python
data_to_scale = data.iloc[:, 5:][data.iloc[:, 5:].sum(axis=1) > 2]
bool(data_to_scale.size)
```

```python
(data.iloc[:, 5:].sum(axis=1) > 1.015).any()
```

```python
warnings.filterwarnings("ignore")
model = keras.models.load_model(model_path)
model.summary()
```

```python
experiment1 = data[(data['ch4_flow']==75) & (data['nc']==10) & (data['sc']==3)
                   & ((data['m_Ni'] - 0.058238972).abs() < 0.000001)]               
experiment2 = data[(data['ch4_flow']==50) & (data['nc']==3) & (data['sc']==3)
                   & ((data['m_Ni'] - 0.291194858).abs() < 0.00001)]
experiment3 = data[(data['ch4_flow']==70) & (data['nc']==1) & (data['sc']==3)
                   & ((data['m_Ni'] - 0.145597429).abs() < 0.00001)]
```

```python
uq_values1 = [experiment1[col].unique() for col in data.columns[:5]]
uq_values2 = [experiment2[col].unique() for col in data.columns[:5]]
uq_values3 = [experiment3[col].unique() for col in data.columns[:5]]
```

```python
e1_T   = np.array(experiment1['t_center'])
e1_ch4 = np.array(experiment1['ch4'])
e1_h2  = np.array(experiment1['h2'])
e1_co2 = np.array(experiment1['co2'])
e1_co  = np.array(experiment1['co'])

e2_T   = np.array(experiment2['t_center'])
e2_ch4 = np.array(experiment2['ch4'])
e2_h2  = np.array(experiment2['h2'])
e2_co2 = np.array(experiment2['co2'])
e2_co  = np.array(experiment2['co'])

e3_T   = np.array(experiment3['t_center'])
e3_ch4 = np.array(experiment3['ch4'])
e3_h2  = np.array(experiment3['h2'])
e3_co2 = np.array(experiment3['co2'])
e3_co  = np.array(experiment3['co'])
```

```python
t_dense1 = np.linspace(e1_T.min(), e1_T.max(), 200)
x_dense1 = experiment1.iloc[0, :5]
X_dense1 = pd.DataFrame(columns=x_dense1.index)
X_dense1 = X_dense1.append([x_dense1] * t_dense1.size, ignore_index=True)
X_dense1['t_center'] = t_dense1

t_dense2 = np.linspace(e2_T.min(), e2_T.max(), 200)
x_dense2 = experiment2.iloc[0, :5]
X_dense2 = pd.DataFrame(columns=x_dense2.index)
X_dense2 = X_dense2.append([x_dense2] * t_dense2.size, ignore_index=True)
X_dense2['t_center'] = t_dense2

t_dense3 = np.linspace(e3_T.min(), e3_T.max(), 200)
x_dense3 = experiment3.iloc[0, :5]
X_dense3 = pd.DataFrame(columns=x_dense3.index)
X_dense3 = X_dense3.append([x_dense3] * t_dense3.size, ignore_index=True)
X_dense3['t_center'] = t_dense3
```

```python
# Load scaler
scaler = joblib.load(scaler_path)
```

```python
X_dense_scaled1 = scaler.transform(X_dense1)
X_dense_scaled2 = scaler.transform(X_dense2)
X_dense_scaled3 = scaler.transform(X_dense3)
```

```python
Y_dense_pred1 = model.predict(X_dense_scaled1)
Y_dense_pred2 = model.predict(X_dense_scaled2)
Y_dense_pred3 = model.predict(X_dense_scaled3)
```

```python
plt.figure(figsize=(14,10))
plt.scatter(e1_T, e1_ch4, s=25, color='blue', label='CH$_4$ (pomiar)');
plt.plot(t_dense1, Y_dense_pred1[:, 0], color = 'blue', label='CH$_4$ (sie)');

plt.scatter(e1_T, e1_h2, s=25, color='red', label='H$_2$ (pomiar)');
plt.plot(t_dense1, Y_dense_pred1[:, 1], color = 'red', label='H$_2$ (sie)');

plt.scatter(e1_T, e1_co2, s=25, color='green', label = 'CO$_2$ (pomiar)');
plt.plot(t_dense1, Y_dense_pred1[:, 2], color = 'green', label='CO$_2$ (sie)');

plt.scatter(e1_T, e1_co, s=25, color='orange', label = 'CO (pomiar)');
plt.plot(t_dense1, Y_dense_pred1[:, 3], color = 'orange', label='CO (sie)');
# \textit{line 2}
plt.xlabel(r'$T_{\rm center}$ [K]', fontsize=18)
plt.ylabel('Zawarto skadnika [-]', fontsize=18)
plt.legend();

fig_number = 1
fig_file = figs_path + "_continuity" + str(fig_number)
plt.savefig(fig_file);
```

```python
plt.figure(figsize=(14,10))
plt.scatter(e2_T, e2_ch4, s=25, color='blue', label='CH$_4$ (pomiar)');
plt.plot(t_dense2, Y_dense_pred2[:, 0], color = 'blue', label='CH$_4$ (sie)');

plt.scatter(e2_T, e2_h2, s=25, color='red', label='H$_2$ (pomiar)');
plt.plot(t_dense2, Y_dense_pred2[:, 1], color = 'red', label='H$_2$ (sie)');

plt.scatter(e2_T, e2_co2, s=25, color='green', label = 'CO$_2$ (pomiar)');
plt.plot(t_dense2, Y_dense_pred2[:, 2], color = 'green', label='CO$_2$ (sie)');

plt.scatter(e2_T, e2_co, s=25, color='orange', label = 'CO (pomiar)');
plt.plot(t_dense2, Y_dense_pred2[:, 3], color = 'orange', label='CO (sie)');
# \textit{line 2}
plt.xlabel(r'$T_{\rm center}$ [K]', fontsize=18)
plt.ylabel('Zawarto skadnika [-]', fontsize=18)
plt.legend();

fig_number = 2
fig_file = figs_path + "_continuity" + str(fig_number)
plt.savefig(fig_file);
```

```python
plt.figure(figsize=(14,10))
plt.scatter(e3_T, e3_ch4, s=25, color='blue', label='CH$_4$ (pomiar)');
plt.plot(t_dense3, Y_dense_pred3[:, 0], color = 'blue', label='CH$_4$ (sie)');

plt.scatter(e3_T, e3_h2, s=25, color='red', label='H$_2$ (pomiar)');
plt.plot(t_dense3, Y_dense_pred3[:, 1], color = 'red', label='H$_2$ (sie)');

plt.scatter(e3_T, e3_co2, s=25, color='green', label = 'CO$_2$ (pomiar)');
plt.plot(t_dense3, Y_dense_pred3[:, 2], color = 'green', label='CO$_2$ (sie)');

plt.scatter(e3_T, e3_co, s=25, color='orange', label = 'CO (pomiar)');
plt.plot(t_dense3, Y_dense_pred3[:, 3], color = 'orange', label='CO (sie)');
# \textit{line 2}
plt.xlabel(r'$T_{\rm center}$ [K]', fontsize=18)
plt.ylabel('Zawarto skadnika [-]', fontsize=18)
plt.legend()

fig_number = 3
fig_file = figs_path + "_continuity" + str(fig_number)
plt.savefig(fig_file);
```
